```python
import pandas as pd
import seaborn as sns

komtopp50 = pd.ExcelFile("../Data/komtopp50_2020.xlsx")
df = pd.read_excel(komtopp50, 'Totalt')

df.rename(columns={df.columns[0]:"Rang 2020", df.columns[1]:"Rang 2019", df.columns[2]:"Kommun", df.columns[3]:"Folkmängd 2020", df.columns[4]:"Folkmängd 2019", df.columns[5]:"Förändring"}, inplace= True)
df = df.drop([0,1,2,3,4,5]).reset_index(drop=True)
df.sort_values("Kommun") 
df

```

```python
sorttopp = df.sort_values("Folkmängd 2020", ascending=False)
sorttopp
```

```python
sorttopp.tail(5)
```

```python
total2019 = df["Folkmängd 2019"].sum()
print(total2019)
total2020 = df["Folkmängd 2020"].sum()
print(total2020)
```

```python
popbigest = df.sort_values("Folkmängd 2020", ascending=False).head()
popbigest
```

```python
popsmall = df.sort_values("Folkmängd 2020", ascending=False).tail()
popsmall
```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(1,2, dpi=100, figsize=(12,4))
sns.barplot(data=popbigest, x="Kommun", y="Folkmängd 2020", ax=ax[0])
sns.barplot(data=popsmall, x="Kommun", y="Folkmängd 2020", ax=ax[1])
```

```python
df1 = pd.read_excel(komtopp50, 'Män')
df1.rename(columns={df1.columns[0]:"Rang 2020", df1.columns[1]:"Rang 2019", df1.columns[2]:"Kommun", df1.columns[3]:"Folkmängd 2020", df1.columns[4]:"Folkmängd 2019", df1.columns[5]:"Förändring"}, inplace= True)
df1 = df1.drop([0,1,2,3,4,5]).reset_index(drop = True)
df1.sort_values("Kommun")
df1['Kön'] ="Män"
df1
```

```python
df2 = pd.read_excel(komtopp50, 'Kvinnor')
df2.rename(columns={df2.columns[0]:"Rang 2020", df2.columns[1]:"Rang 2019", df2.columns[2]:"Kommun", df2.columns[3]:"Folkmängd 2020", df2.columns[4]:"Folkmängd 2019", df2.columns[5]:"Förändring"}, inplace= True)
df2 = df2.drop([0,1,2,3,4,5]).reset_index(drop = True)

df2.sort_values("Kommun")
df2["Kön"]="Kvinnor"
df2
```

```python
alldata = df1.append(df2, ignore_index=True)
alldata=alldata.set_index("Kommun")
alldata=alldata.drop(columns=['Rang 2020','Rang 2019'])
alldata
```
